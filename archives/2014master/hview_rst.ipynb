```python
%matplotlib inline

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

import romspy

ncfile = '/Users/teruhisa/mnt/apps/OB500_fennelP/NL04/ob500_rst.nc'
mapfile = '/Users/teruhisa/Dropbox/Data/deg_OsakaBayMap_okada.bln'

def hview(name):
    fig, ax = plt.subplots(1,1,figsize=[7,6])
    k = 20
    interval=None
    if 'N' in name:
        interval=np.linspace(0,10,10)
    elif 'P' in name:
        interval=np.linspace(0,1,10)
    elif name == 'salt':
        interval=np.linspace(23,33,10)
    elif name == 'chlorophyll':
        interval=np.linspace(0,400,10)
    elif name == 'oxygen':
        k=1
    ax = romspy.hview(ncfile, name, time, k, mapfile=mapfile, cblabel=name, interval=interval)

romspy.get_time(ncfile)
```
