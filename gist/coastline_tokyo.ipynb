```python
# (c) 2015 Teruhisa Okada

%matplotlib inline

import matplotlib.pyplot as plt
import shapefile as sf

def coastline(shpname):
    map = sf.Reader(shpname)
    meta = map.records()
    shapes = map.shapes()
    for n in range(len(meta)):
        lon = [px[0] for px in shapes[n].points]
        lat = [px[1] for px in shapes[n].points]
        plt.plot(lon, lat, 'k-',ms=2)
```

```python
fig, ax = plt.subplots(figsize=(10,12))

coastline('Z:/Dropbox/Maps/gm-jpn/gm-jpn-all_u_2/coastl_jpn')
#coastline('/Users/teruhisa/Dropbox/Maps/gm-jpn/gm-jpn-all_u_2/inwatera_jpn')

plt.xlim(139.60, 140.15)
plt.ylim(35.15, 35.70)

plt.savefig('coastline_tokyo.png', bbox_inches='tight', dpi=300)
```
