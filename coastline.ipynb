```python
%matplotlib inline

import matplotlib.pyplot as plt
import shapefile as sf

def coastline(shpname):
    map = sf.Reader(shpname)
    meta = map.records()
    shapes = map.shapes()
    for n in range(len(meta)):
        lon = [px[0] for px in shapes[n].points]
        lat = [px[1] for px in shapes[n].points]
        plt.plot(lon, lat, 'k-',ms=2)
```

```python
from mpl_toolkits.basemap import Basemap, cm
import numpy as np

m = Basemap(llcrnrlon=125.,llcrnrlat=25.,urcrnrlon=153.,urcrnrlat=47.,
            projection='lcc', lat_1=20., lon_0=135.,
            resolution ='l',area_thresh=1000.)

m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='#99ffff')
#m.fillcontinents(color='#cc9966',lake_color='#99ffff')
m.fillcontinents(color='w', lake_color='#99ffff')

m.drawparallels(np.arange(30,50,10),labels=[1,1])
m.drawmeridians(np.arange(130,160,10),labels=[1,1,1])

plt.savefig('coastline_japan.png', dpi=600, bbox_inches='tight')
```
