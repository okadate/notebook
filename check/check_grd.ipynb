```python
%matplotlib inline

import matplotlib.pyplot as plt
import netCDF4
import numpy as np

def check_grd(ncfile):
    nc = netCDF4.Dataset(ncfile, 'r')
    h = nc.variables['h'][:,:]
    fig, ax = plt.subplots(1,1)
    pc = ax.pcolor(h, vmin=0, vmax=110)
    plt.colorbar(pc)
    #ax.set_xlim(100,120)
    #ax.set_ylim(100,130)
    ax.set_xlim(10,50)
    ax.set_ylim(80,110)

ncfile = '/Users/teruhisa/Dropbox/Data/ob500_grd-9.nc'
check_grd(ncfile)
```
