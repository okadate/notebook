```python
%matplotlib inline

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

import romspy

mapfile = '/Users/teruhisa/Dropbox/Data/deg_OsakaBayMap_okada.bln'

ncfile = '/Users/teruhisa/Dropbox/Data/ob500_grd-8.nc'


fig, ax = plt.subplots(1,1,figsize=[7,6])
ax = romspy.hview(ncfile, 'h', mapfile=mapfile, cmap='ocean', cff=-1, interval=np.arange(-100,1,20))

def names():
    plt.text(135.22,34.55,'Osaka Bay')
    plt.text(135.44,34.7,'Yodo R.', backgroundcolor='1.0')
    plt.text(135.42,34.6,'Yamato R.', backgroundcolor='1.0')
    plt.text(134.97,34.66,'Akashi Strait')
    plt.text(135.1,34.28,'Kitan Strait')

names()
plt.savefig('names.png', bbox_inches='tight', dpi=300)
```

```python
import pandas as pd

fig, ax = plt.subplots(1,1,figsize=[5.8,6])
romspy.basemap(mapfile)


def stations13(csvfile):
    df = pd.read_csv(csvfile, index_col='station')
    for sta in df.index:
        plt.text(df.lon[sta]-0.02, df.lat[sta], '{}'.format(sta), size=15, ha='right', va='center')
        if sta in [3,4,5,6,12,13]:
            plt.plot(df.lon[sta], df.lat[sta], 'rD', ms=13)
        else:
            plt.plot(df.lon[sta], df.lat[sta], 'bD', ms=13)

stations13('/Users/teruhisa/Dropbox/Data/stations13.csv')
plt.title('observation stations')
plt.savefig('stations13.png', bbox_inches='tight', dpi=300)
```
