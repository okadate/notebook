```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

K_NO3 = 1.0/( 0.02 /14.01*1000.0) #1.4
K_NH4 = 1.0/( 0.02 /14.01*1000.0) #1.41.4
K_PO4 = 1.0/( 0.01 /30.97*1000.0) #33.333

NO3 = np.linspace(0,5,50)
NH4 = np.linspace(0,5,50)
PO4 = np.linspace(0,0.5,50)

cff1=NH4*K_NH4
cff2=NO3*K_NO3
inhNH4=1.0/(1.0+cff1)
L_NH4=cff1/(1.0+cff1)
L_NO3=cff2*inhNH4/(1.0+cff2)
LTOT=L_NO3+L_NH4

cff3=PO4*K_PO4
L_PO4=cff3/(1.0+cff3)

fig, ax = plt.subplots(1,2,figsize=[12,5])
ax[0].plot(NO3*14.01/1000.0, L_NO3, 'g-', label='NO3')
ax[0].plot(NH4*14.01/1000.0, L_NH4, 'b-', label='NH4')
ax[0].plot(NH4*14.01/1000.0, LTOT, 'k-', label='Total')
ax[1].plot(PO4*30.97/1000.0, L_PO4, 'r-', label='PO4')
ax[0].set_xlim(0,0.1)
ax[1].set_xlim(0,0.03)
for i in range(2):
    ax[i].legend()
    ax[i].grid()
    ax[i].set_ylim(0,1)
```
