```python
% matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import romspy

#start = [34.692446, 135.447326]
start = [34.691690, 135.447950]
end = [34.649289, 135.317306]

lon = np.linspace(start[1], end[1], 21)
lat = np.linspace(start[0], end[0], 21)

fig, ax = plt.subplots(1,1,figsize=[7,7])
romspy.basemap('F:/okada/notebook/deg_OsakaBayMap_okada.bln')
ax = plt.gca()
ax.plot(lon, lat)
```

```python
ncfile = 'Z:/apps/OB500P/case13/ob500_avg.nc'
mapfile = 'F:/okada/notebook/deg_OsakaBayMap_okada.bln'

fig, ax = plt.subplots(1,1,figsize=[8,7])

grd = romspy.hview.Dataset(ncfile)#, mapfile=mapfile)
grd.hview('salt', time=7, k=20)

ax = plt.gca()
ax.plot(lon, lat, 'w.-')
ax.set_xlim(135.3, 135.5)
ax.set_ylim(34.6, 34.75)
```
