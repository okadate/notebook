```python
%matplotlib inline

import matplotlib.pyplot as plt
import netCDF4
import numpy as np

def check_grd(ncfile):
    nc = netCDF4.Dataset(ncfile, 'r')
    mask_u = nc.variables['mask_u'][:,:]
    mask_v = nc.variables['mask_v'][:,:]
    mask_psi = nc.variables['mask_psi'][:,:]
    fig, ax = plt.subplots(1,3,figsize=[15,5])
    ax[0].pcolor(mask_u)
    ax[1].pcolor(mask_v)
    ax[2].pcolor(mask_psi)

ncfile = '/Users/teruhisa/Dropbox/Data/ob500_grd-8.nc'
check_grd(ncfile)
```
