```python
%matplotlib inline

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

import romspy

#ncfile = '/Users/teruhisa/mnt/apps/OB500_fennelP/NL05/ob500_avg.nc'
#ncfile = '/Users/teruhisa/mnt/mpi00/Bio-4.2.1-201208/NL/osaka-bay_his.nc'
ncfile = '/Users/teruhisa/mnt/mpi00/Bio-NL-4.2.1-201208/his/osaka-bay_his_0169.nc'
grdfile = '/Users/teruhisa/Dropbox/Data/osaka-bay_grd_v1.2.nc'
mapfile = '/Users/teruhisa/Dropbox/Data/deg_OsakaBayMap_okada.bln'

def hview(name, interval=None, k=20, fmt='%i', cmap='jet'):
    fig, ax = plt.subplots(1,1,figsize=[7,6])
    if name == 'oxygen':
        cblabel = u'Dissolved oxygen [mmol O2 m$^{-3}$]'
    else:
        cblabel = name
    ax = romspy.hview(ncfile, name, time, k, mapfile=mapfile, cblabel=cblabel, interval=interval, fmt=fmt, grdfile=grdfile, cmap=cmap)

romspy.get_time(ncfile, 'all')
#romspy.get_variables(ncfile)
```
