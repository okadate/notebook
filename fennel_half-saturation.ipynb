```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

K_NO3 = 1.0/( 0.02 /14.01*1000.0) #1.4
K_NH4 = 1.0/( 0.02 /14.01*1000.0) #1.41.4
K_PO4 = 1.0/( 0.002 /30.97*1000.0) #33.333

NO3 = np.linspace(0,5,50)
NH4 = np.linspace(0,5,50)
PO4 = np.linspace(0,5,50)

cff1=NH4*K_NH4
cff2=NO3*K_NO3
inhNH4=1.0/(1.0+cff1)
L_NH4=cff1/(1.0+cff1)
L_NO3=cff2*inhNH4/(1.0+cff2)
LTOT=L_NO3+L_NH4

cff3=PO4*K_PO4
L_PO4=cff3/(1.0+cff3)

fig, ax = plt.subplots(1,3,figsize=[12,5])
ax[0].plot(NO3, L_NO3, 'g-')
ax[1].plot(NH4, LTOT, 'b-')
ax[2].plot(PO4, L_PO4, 'r-')
```
