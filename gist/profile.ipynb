```python
# (c) 2015 Teruhisa Okada

% matplotlib inline

import matplotlib.pyplot as plt
import datetime
import romspy

obs = 'F:/okada/Dropbox/Data/ob500_obs_2012_obweb-4.nc'
sta = 'Z:/roms/Apps/OB500_fennelP/NL12/ob500_sta.nc'

vname='temp'
station=3

fig, ax = plt.subplots(1,1,figsize=[10,5])
p = romspy.profile.Profile(obsfile=obs, freefile=sta)
dates = [datetime.datetime(2012,8,1,0), datetime.datetime(2012,9,1,0)]

p.plot(ax, vname, station, dates)
ax.set_xlim(5,30)

#p.savefig('profile_temp_3_aug.png')
```

```python
vname='temp'
station=3

for month in range(1,13):
    fig, ax = plt.subplots(1,1,figsize=[10,5])
    p = romspy.profile.Profile(obsfile=obs, freefile=sta)
    #p = romspy.profile.Profile(freefile=sta)
    if month == 12:
        dates = [datetime.datetime(2012,12,1,0), datetime.datetime(2013,1,1,0)]
    else:
        dates = [datetime.datetime(2012,month,1,0), datetime.datetime(2012,month+1,1,0)]
    p.plot(ax, vname, station, dates)
    ax.set_xlim(5,30)
    ax.set_title('2012/{}'.format(month))
```
