```python
# (c) 2015 Teruhisa Okada

% matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import romspy

ncfile = 'Z:/roms/Apps/OB500_fennelP/NL12/ob500_avg.nc'
mapfile = 'Z:/Data/deg_OsakaBayMap_okada.bln'

dia = romspy.hview.Dataset(ncfile, mapfile=mapfile)
dia.print_time('all')
dia.print_varname(ndim=4)
```

```python
fig, ax = plt.subplots(1,1,figsize=[8,7])

dia.hview('oxygen', time=6, k=1, fmt='%i', cff=romspy.mol2g_O2, interval=np.linspace(0,10,11))
#dia.savefig('hview_avg_salt_sur.png')
dia.show()
```
