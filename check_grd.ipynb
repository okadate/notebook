```python
%matplotlib inline

import matplotlib.pyplot as plt
import netCDF4
import numpy as np

def check_grd(ncfile):
    nc = netCDF4.Dataset(ncfile, 'r')
    h = nc.variables['h'][:,:]
    fig, ax = plt.subplots(1,1)
    #pc = ax.contourf(h, np.arange(1, 20, 1), extend='max')
    pc = ax.pcolor(h, vmin=0.5, vmax=1)
    plt.colorbar(pc)
    ax.set_xlim(50,120)
    ax.set_ylim(50,120)

ncfile = '/Users/teruhisa/Dropbox/Data/ob500_grd-6.nc'
check_grd(ncfile)
```
