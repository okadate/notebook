```python
%matplotlib inline

import matplotlib.pyplot as plt
import netCDF4

ncfile = '/Users/teruhisa/Dropbox/Data/ob500_bry_2012_fennelP-1.nc'
nc = netCDF4.Dataset(ncfile, 'r')

def check_zeta():
    zeta_west = nc.variables['zeta_west']
    zeta_south = nc.variables['zeta_south']
    #print zeta_south
    fig, ax = plt.subplots(2,1)
    ax[0].plot(zeta_west[:1000, 100])
    ax[1].plot(zeta_south[:1000, 100])

def check3d(name):
    var_west = nc.variables[name+'_west']
    var_south = nc.variables[name+'_south']
    fig, ax = plt.subplots(2,2, figsize=[13,3])
    ax[0][0].plot(var_west[:, 19, 100])
    ax[1][0].plot(var_west[:, 0, 100])
    ax[0][1].plot(var_south[:, 19, 100])
    ax[1][1].plot(var_south[:, 0, 100])
    ax[0][0].set_title('west surface '+name)
    ax[1][0].set_title('west bottom '+name)
    ax[0][1].set_title('south surface '+name)
    ax[1][1].set_title('south bottom '+name)

check_zeta()
```
