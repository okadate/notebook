#Time series of rain and river


```python
%matplotlib inline
import netCDF4
import matplotlib.pyplot as plt

riverfile = '/Users/teruhisa/mpi00/fujii/Data/osaka-bay_river_201206-2_LQ2.nc'
river = netCDF4.Dataset(riverfile, 'r')
river_time = river.variables['river_time'][:]
yodo = river.variables['river_transport'][:,6]
yamato = river.variables['river_transport'][:,11]
```

```python
rainfile = '/Users/teruhisa/mpi00/fujii/Data/osaka-bay_rain_201208.nc'
rain_nc = netCDF4.Dataset(rainfile,'r')
rain_time = rain_nc.variables['rain_time'][:]
rain = rain_nc.variables['rain'][:,0,0]
```

```python
fig, ax = plt.subplots(1,1,figsize=[10,1.5])

ax.plot(river_time-60, -yodo*2, 'r-', label='Yodo R.')
ax.plot(river_time-60, -yamato*2, 'g-', label='Yamato R.')
ax.set_xlim(0,31)
ax.set_ylim(0,1000)
ax.set_xticks(range(0,31,4))
#ax.set_xlabel('August, 2012')
ax.set_ylabel("River discharge [m s$^{-1}$]")
ax.grid()
ax.legend(loc=1)

plt.tick_params(labelbottom='off')

#ax2 = ax.twinx()
#ax2.bar(rain_time-61, -rain, align='center', width=0.03, alpha=0.7, label='Osaka')
#ax2.set_ylim(-0.006, 0)
#ax2.set_ylabel('Rain fall rate [kg m$^{2}$ s$^{-1}$]')
#ax2.legend(loc=1)
```
