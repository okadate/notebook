```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure()
list = []
x = np.arange(0, 10, 0.1)

for a in range(0, 10):
    y = np.sin(x + a)
    frame = plt.plot(x, y)
    list.append(frame)

ani = animation.ArtistAnimation(fig, list)
ani.save("output.gif")
```
