```python
%matplotlib inline

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

import romspy

ncfile = '/Users/teruhisa/mnt/hydcl00/OB500/Biology_test_0/test_boundary/ob500_avg.nc'
mapfile = '/Users/teruhisa/Dropbox/Data/deg_OsakaBayMap_okada.bln'

def hview(name, k=20):
    fig, ax = plt.subplots(1,1,figsize=[7,6])
    if 'N' in name:
        interval=np.linspace(0,10,10)
    elif 'P' in name:
        interval=np.linspace(0,1,10)
    elif name == 'salt':
        interval=np.linspace(23,33,10)
    elif name == 'chlorophyll':
        interval=np.linspace(0,20,10)
    else:
        interval=None
    ax = romspy.hview(ncfile, name, time, k, mapfile=mapfile, cblabel=name, interval=interval)

romspy.get_time(ncfile, 'all')
```
