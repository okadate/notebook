```python
%matplotlib inline

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

import romspy

ncfile = '/Users/teruhisa/mnt/apps/OB500_fennelP/NL05/ob500_avg.nc'
mapfile = '/Users/teruhisa/Dropbox/Data/deg_OsakaBayMap_okada.bln'

def hview(name, interval=None, k=20, fmt='%i'):
    fig, ax = plt.subplots(1,1,figsize=[7,6])
    ax = romspy.hview(ncfile, name, time, k, mapfile=mapfile, cblabel=name, interval=interval, fmt=fmt)

romspy.get_time(ncfile, 'all')
```

```python
time = dt.datetime(2012, 2, 15, 18)

hview('temp')
hview('salt', np.linspace(23,33,11))
hview('oxygen', np.linspace(140,300,17))
hview('oxygen', np.linspace(140,300,17), k=1)
```

```python
fmt = '%1.1f'
hview('chlorophyll', np.linspace(0,5,11), fmt=fmt)
hview('phytoplankton', np.linspace(0,2,11), fmt=fmt)
hview('zooplankton', np.linspace(0,2,11), fmt=fmt)
```

```python
interval = np.linspace(0,20,11)
hview('NH4', interval)
hview('NO3', interval)
hview('LdetritusN', interval)
hview('SdetritusN', interval)
```

```python
interval = np.linspace(0,1,11)
fmt = '%1.1f'
hview('PO4', interval, fmt=fmt)
hview('LdetritusP', interval, fmt=fmt)
hview('SdetritusP', interval, fmt=fmt)
```
