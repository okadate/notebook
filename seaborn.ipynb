```python
% matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

x = np.random.normal(size=100) #ランダムデータをnumpy arrayとして作る

titanic = sns.load_dataset("titanic") ##kaggleで有名な、タイタニック号の生死者データ
tips = sns.load_dataset("tips")  ## お店の食事時間と会計総額とチップの関係のデータ
iris = sns.load_dataset("iris")  ## Rでお馴染みのアヤメの統計データ
```

```python
sns.distplot(x, kde=False, rug=False, bins=10)
```

```python
sns.distplot(x)
```

```python
sns.distplot(x, rug=False)
```

```python
sns.jointplot('sepal_width', 'petal_length', data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.barplot(x='sex', y='survived', data=titanic, hue='class')
```

```python
titanic_grpby = titanic.groupby( ['sex', 'class'])
titanic_data_for_graph = titanic_grpby['survived'].aggregate(sum).reset_index()

sns.barplot(x='sex', y='survived', hue= 'class', data=titanic_data_for_graph)
```
