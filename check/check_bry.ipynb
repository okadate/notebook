```python
%matplotlib inline

import matplotlib.pyplot as plt
import netCDF4

import romspy

ncfile1 = 'F:/okada/Dropbox/Data/ob500_bry_2012_fennelP-4.nc'
ncfile2 = 'F:/okada/Dropbox/Data/ob500_bry_2012_fennelP-5.nc'

#romspy.get_time(ncfile, tunit=romspy.JST, name='time_hourly')
#romspy.get_variables(ncfile)

def check_zeta(ncfile):
    nc = netCDF4.Dataset(ncfile, 'r')
    zeta_west = nc.variables['zeta_west']
    zeta_south = nc.variables['zeta_south']
    #print zeta_south
    fig, ax = plt.subplots(2,1)
    ax[0].plot(zeta_west[:1000, 100])
    ax[1].plot(zeta_south[:1000, 100])

check_zeta(ncfile1)
check_zeta(ncfile2)
```

```python
def check3d(ncfile, name):
    nc = netCDF4.Dataset(ncfile, 'r')
    tname = nc.variables[name+'_west'].time
    time = nc.variables[tname][:]
    time = netCDF4.num2date(time, romspy.JST_days)
    var_west = nc.variables[name+'_west']
    var_south = nc.variables[name+'_south']
    fig, ax = plt.subplots(1,2, figsize=[13,3])
    ax[0].plot(time, var_west[:, 19, 100], label='surface')
    ax[0].plot(time, var_west[:, 0, 100], label='bottom')
    ax[1].plot(time, var_south[:, 19, 100], label='surface')
    ax[1].plot(time, var_south[:, 0, 100], label='bottom')
    ax[0].set_title('{}_west'.format(name))
    ax[1].set_title('{}_south'.format(name))
    ax[1].legend()

def plot(vname):
    check3d(ncfile1, vname)
    check3d(ncfile2, vname)

plot('salt')
```
