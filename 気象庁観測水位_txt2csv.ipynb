```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

data = '/Users/teruhisa/Dropbox/Data/hry201208TK.txt'

f = open(data)

zeta = []
for line in f.readlines():
    for t in range(24):
        zeta.append(int(line[t*3:t*3+3]))

zeta = pd.Series(data=zeta)
zeta.plot()
zeta.to_csv(data[:-3]+'csv')
```
